```python
def flattenShootingData(shootingData):
  try:
    # key
    mainKeyIndex = 5 
    # multistack barchart
    fg2aIndex = 13
    fg3aIndex = 17
    fg3pctIndex = 18
    fg2pctIndex = 14
    # piechart 1
    fgaFreqIndex = 6
    # piechart 2
    fg2aFregIndex = 11
    fg3aFreqIndex = 15

    flattenData = []
    for data in shootingData['data']:
      flattenData.append({
        "key": data[mainKeyIndex],
        "FG2A": data[fg2aIndex],
        "FG3A": data[fg3aIndex],
        "FG2_PCT": data[fg2pctIndex],
        "FG3_PCT": data[fg3pctIndex],
        "FGA_FREQ": data[fgaFreqIndex],
        "FG2A_FREQ": data[fg2aFregIndex],
        "FG3A_FREQ": data[fg3aFreqIndex],
      })
    return flattenData
  except Exception as e:
    print(e)
```
